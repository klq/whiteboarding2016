```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy

def weird_result(candidate_number):
    data_numbers = [2, 7, 1, 5, 10]
    return sum([ (candidate_number-d)**2 for d in data_numbers])

result_list = []
for candidate_number in numpy.arange(0,10.1,0.1):
    result_list.append(weird_result(candidate_number))
print 'The minimum result is: ', min(result_list)
```

```python
plt.plot(numpy.arange(0,10.1,0.1), result_list)

plt.xlabel('Candidate Number')
plt.ylabel('Result')
plt.title('Now it\'s more obvious, isn\'t it?')
plt.grid(True)
plt.savefig("candidate_number.png")
plt.show()
```
